```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
filename = 'C:/Users/Tathagat/Documents/Images/IMG_6242.jpg'
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename, target_size = (224,224))
```

```python
plt.imshow(img)
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
resized_img = image.img_to_array(img)
final_image = np.expand_dims(resized_img, axis=0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
predictions = mobile.predict(final_image)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results = imagenet_utils.decode_predictions(predictions)
```

```python
print(results)
```

```python
# The five predictions made for the image is 'warplane' with 65.9% probability, 'wing' with 18.2% probability, 
# 'airliner' with 9.9% probability, 'projectile' with 2.8% probability, and 'missile' with 1.5% probability
```
