```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.applications import imagenet_utils
```

```python
filename = 'C:/Users/Tathagat/Documents/Images/IMG_6082.jpg'
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename, target_size = (224,224))
```

```python
plt.imshow(img)
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
resized_img = image.img_to_array(img)
final_image = np.expand_dims(resized_img, axis=0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
predictions = mobile.predict(final_image)
```

```python
results = imagenet_utils.decode_predictions(predictions)
```

```python
print(results)
```

```python
# The five predictions made for the image is 'tank' with 99.97% probability, 'cannon' with 0.019% probability, then 
# 'half_track', then 'plow', then 'harvester' with much smaller probabilities
```
